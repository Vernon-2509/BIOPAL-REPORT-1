-Trend analysis (2007–2018) revealed stability across most variables. BMI rose slightly, while sleep and depression scores remained low but unchanged.

-The correlation matrix showed weak relationships, supporting the idea that depression has no single cause but is influenced by multiple subtle factors.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# -----------------------------
# Step 0: Define years and NHANES cycles
# -----------------------------
years = ["2007","2009","2011", "2013", "2015", "2017"]
cycles = ["E","F","G", "H", "I", "J"]

# -----------------------------
# Step 1: Load All Raw Datasets
# -----------------------------
print("Loading all raw datasets...")

def load_data(file_prefix, columns):
    """Helper function to load and concatenate data across cycles."""
    urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/{file_prefix}_{c}.xpt" for y, c in zip(years, cycles)]
    df_list = []
    for url in urls:
        try:
            df = pd.read_sas(url)
            # Ensure we only select columns that actually exist in the file
            existing_cols = [col for col in columns if col in df.columns]
            df_list.append(df[existing_cols])
        except Exception as e:
            print(f"Could not load or process {url}. Error: {e}")
    return pd.concat(df_list, ignore_index=True)
#
df_demo = load_data("DEMO", ["SEQN","SDDSRVYR", "RIAGENDR", "RIDAGEYR", "RIDRETH1"])
df_bmi = load_data("BMX", ["SEQN", "SDDSRVYR","BMXBMI"])
df_smoking = load_data("SMQ", ["SEQN", "SDDSRVYR","SMQ040"])
df_physical = load_data("PAQ", ["SEQN", "SDDSRVYR","PAD680"])
df_sleep = load_data("SLQ", ["SEQN", "SDDSRVYR","SLD010H"])

# Define the columns you want from the DPQ dataset
dpq_columns = ["SEQN", "SDDSRVYR", "DPQ010", "DPQ020", "DPQ030", "DPQ040", "DPQ050"]

# Load the dataset (you must have a function like load_data implemented)
df_depression = load_data("DPQ", dpq_columns)

# Replace codes 7 (Refused) and 9 (Don't know) with NaN
dpq_question_cols = ["DPQ010", "DPQ020", "DPQ030", "DPQ040", "DPQ050"]
df_depression[dpq_question_cols] = df_depression[dpq_question_cols].replace([7, 9], np.nan)

# Optional: convert to float in case values are stored as strings
df_depression[dpq_question_cols] = df_depression[dpq_question_cols].astype(float)

# Calculate PHQ-9 partial total score (based on 5 questions here)
df_depression["DPQ_total_score"] = df_depression[dpq_question_cols].sum(axis=1)

# Drop individual DPQ question columns
df_depression.drop(columns=dpq_question_cols, inplace=True)

# Preview result
df_depression.head()

df_pulse = load_data("BPX", ["SEQN", "SDDSRVYR","BPXPLS","BPXSY1","BPXSY2","BPXSY3"])
df_pulse["BPXSY"]=(df_pulse["BPXSY1"]+df_pulse["BPXSY2"]+df_pulse["BPXSY3"])/3
df_pulse=df_pulse.drop(columns=["BPXSY1","BPXSY2","BPXSY3"])
```

```python
df=df_demo.merge(df_bmi,on=["SEQN"],how="left")
df=df.merge(df_smoking,on=["SEQN"],how="left")
df=df.merge(df_physical,on=["SEQN"],how="left")
df=df.merge(df_sleep,on=["SEQN"],how="left")
df=df.merge(df_depression,on=["SEQN"],how="left")
df=df.merge(df_pulse,on=["SEQN"],how="left")
df.head()
```

```python
import pandas as pd
import numpy as np

# --- Smoking status: SMQ040 ---
# Replace 7 = Refused and 9 = Don't know with NaN
df["SMQ040"] = df["SMQ040"].replace([7, 9], np.nan)

# Map the smoking status codes to categories
df["SMQ040"] = df["SMQ040"].map({
    1: 'heavy smoker',
    2: 'light smoker',
    3: 'never smoked'
})

# --- Sedentary behavior: PAD680 ---
def categorize_sedentary(mins):
    if pd.isna(mins):
        return "missing"
    elif mins in [7777, 9999]:
        return "unknown"
    elif mins < 240:
        return "low"         # < 4 hours
    elif mins < 480:
        return "moderate"    # 4–8 hours
    elif mins <= 1200:
        return "high"        # >8 hours
    else:
        return "invalid"     # >1200 mins (20 hours), invalid

# Apply categorization function
df["sedentary_category"] = df["PAD680"].apply(categorize_sedentary)

# --- Sleep hours: SLD010H ---
# Replace 7 = Refused and 9 = Don't know with NaN
df["SLD010H"] = df["SLD010H"].replace([7, 9], np.nan)
#replace sleep hours more than 24 with 24
df.loc[df["SLD010H"] > 24, "SLD010H"] = 24
#restrict the study to people 10-35
df = df[(df["RIDAGEYR"] >= 10) & (df["RIDAGEYR"] <= 35)]
```

# ***8. Annual Trends***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import linregress

# Set plot style
sns.set_style("whitegrid", {'grid.linestyle': ':', 'axes.edgecolor': '0.3'})
plt.rcParams.update({
    'font.size': 14,
    'axes.titlesize': 16,
    'axes.titleweight': 'bold',
    'axes.titlepad': 15
})

# Define variables and their human-readable names
variables = {
    'RIDAGEYR': 'Average Age (years)',
    'BMXBMI': 'Median BMI',
    'PAD680': 'Median Sedentary Time (min)',
    'SLD010H': 'Median Sleep Duration (hours)',
    'DPQ_total_score': 'Median Depression Score',
    'BPXPLS': 'Median Pulse Rate (bpm)',
    'BPXSY': 'Median Systolic BP (mmHg)'
}

# Create individual plots for each variable
for var, title in variables.items():
    # Skip missing data for the current variable
    if df[var].isna().all():
        print(f"Skipping {var} (all values missing)")
        continue

    # Create figure
    plt.figure(figsize=(10, 6))

    # Calculate annual values (handle NaNs)
    if var == 'RIDAGEYR':
        # Use non-missing values for age
        annual_data = df.dropna(subset=[var]).groupby('SDDSRVYR')[var].mean()
    else:
        # Use non-missing values for other variables
        annual_data = df.dropna(subset=[var]).groupby('SDDSRVYR')[var].median()

    # Convert to DataFrame for plotting
    annual_df = annual_data.reset_index()

    # Skip if no data remains after dropping NaNs
    if annual_df.empty:
        print(f"No data for {var} after removing NaNs")
        continue

    # Plot with enhanced markers
    plt.plot(
        annual_df['SDDSRVYR'],
        annual_df[var],
        marker='o',
        markersize=10,
        markeredgecolor='white',
        markeredgewidth=1.5,
        linewidth=3,
        color='#1f77b4'
    )

    # Add data point labels
    for idx, row in annual_df.iterrows():
        plt.text(row['SDDSRVYR'], row[var] * 1.02, f'{row[var]:.1f}',
                 ha='center', va='bottom', fontsize=12, fontweight='bold')

    # Add trendline only if sufficient data points exist
    if len(annual_df) >= 2:
        slope, intercept, r_value, p_value, std_err = linregress(
            annual_df['SDDSRVYR'], annual_df[var]
        )

        if p_value < 0.1:  # Show trendline if p<0.1
            plt.plot(
                annual_df['SDDSRVYR'],
                intercept + slope * annual_df['SDDSRVYR'],
                '--',
                color='#ff7f0e',
                linewidth=2.5,
                alpha=0.8
            )
            # Add trend annotation
            direction = "↑ Increasing" if slope > 0 else "↓ Decreasing"
            plt.text(0.05, 0.92,
                     f"Trend: {direction}\n(p = {p_value:.3f})",
                     transform=plt.gca().transAxes,
                     fontsize=12,
                     bbox=dict(facecolor='white', alpha=0.8, edgecolor='#ff7f0e'))
    else:
        print(f"Insufficient data points for trendline in {var}")

    # Customize plot
    plt.title(f'{title} Trends: NHANES 2007-2018', fontsize=16, pad=15)
    plt.xlabel('Survey Year', fontsize=14)
    plt.ylabel(title, fontsize=14)

    # Set x-ticks to actual years
    years = sorted(df['SDDSRVYR'].unique())
    plt.xticks(years, [f"{int(year)}" for year in years])

    # Add grid lines
    plt.grid(True, linestyle=':', alpha=0.7)

    # Adjust y-limits based on variable
    y_min = annual_df[var].min() * 0.95
    y_max = annual_df[var].max() * 1.15

    # Special handling for specific variables
    if var == 'PAD680':
        plt.ylim(0, min(800, y_max))  # Cap at 800 minutes
        plt.yticks(np.arange(0, 801, 100))
    elif var == 'SLD010H':
        plt.ylim(6, 9)  # Focus on typical sleep range
    else:
        plt.ylim(y_min, y_max)

    # Save and show
    plt.tight_layout()
    plt.savefig(f'{var}_annual_trend.png', dpi=300, bbox_inches='tight')
    plt.show()

# Generate trend summary table (with NaN handling)
trend_results = []
for var, title in variables.items():
    # Handle missing data
    clean_df = df.dropna(subset=[var, 'SDDSRVYR'])
    if clean_df.empty:
        trend_results.append({
            'Variable': title,
            'Slope': np.nan,
            'P-value': np.nan,
            'Trend Direction': 'Insufficient Data',
            'Significance': 'N/A'
        })
        continue

    # Calculate annual values
    if var == 'RIDAGEYR':
        annual_data = clean_df.groupby('SDDSRVYR')[var].mean()
    else:
        annual_data = clean_df.groupby('SDDSRVYR')[var].median()

    # Skip regression if <2 data points
    if len(annual_data) < 2:
        trend_results.append({
            'Variable': title,
            'Slope': np.nan,
            'P-value': np.nan,
            'Trend Direction': 'Insufficient Data',
            'Significance': 'N/A'
        })
        continue

    # Perform regression
    years = annual_data.index
    values = annual_data.values
    slope, intercept, r_value, p_value, std_err = linregress(years, values)

    trend_results.append({
        'Variable': title,
        'Slope': slope,
        'P-value': p_value,
        'Trend Direction': 'Increase' if slope > 0 else 'Decrease',
        'Significance': 'Significant' if p_value < 0.05 else 'Not Significant'
    })

trend_df = pd.DataFrame(trend_results)
print("\nAnnual Trend Analysis Summary:")
display(trend_df)
```
